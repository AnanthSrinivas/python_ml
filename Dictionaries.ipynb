**zip**

- Main moto of zipping is we want to combine 2 iterators.

```python
names = ['Ananth','Sai','Srinivas']
age = [25,24,23]

for i,j in zip(names,age):
    print(i,j)
```

```python
names = [50,33,22]
age = [25,24,23]

for i,j in zip(names,age):
    print(i+j)
```

```python
names = [50,33,22]
age = [25,24,23]
h = [5,5,5]
for i,j,k in zip(names,age,h):
    print(i+j+k)

print(list(zip(names,age,h)))
```
